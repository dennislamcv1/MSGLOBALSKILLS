```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

### Exercise 1: Familiarize Yourself with the New Dataset

```python
df = pd.read_csv("ButterflyChartLab_Dataset.csv")
```

```python
df
```

```python
df.info()
```

```python
df["EnglishProductName"].str.contains("Frame").sum()
```

```python
df1 = df[df["EnglishProductName"].str.contains("Frame")]
```

```python
df1
```

```python
df1.reset_index(inplace=True, drop=True)
```

```python
df1
```

### Exercise 2: Create a PivotTable

```python
df2 = pd.pivot_table(data=df1, index=["EnglishProductName"], columns="SalesTerritoryRegion", values=["SalesTotal"], aggfunc='sum')
```

```python
df2
```

```python
df2.columns
```

```python
df2.columns = df2.columns.get_level_values(1)
```

```python
df2.columns
```

```python
df2
```

### Exercise 3: Add a Heatmap to Your Pivot Table

```python
plt.figure(figsize=(15,25))
sns.heatmap(data=df2, cmap="coolwarm", annot=True, fmt='.9g')
plt.show()
```

### Exercise 4: Turn Your Pivot-Heatmap into a Butterfly Combo Chart

```python
sns.jointplot(x=df2.Canada, y=df2.Australia, data=df2)
plt.show()
```
